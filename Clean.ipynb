```julia
mutable struct Partie
    balle::Balle
    trou::Trou
    obstacles::Array{LineSegment,1}
    nb_coups::Int
    is_over::Bool
end
```

```julia
function impulsionBalle!(balle::Balle, new_v::Float64, new_theta::Float64)
    balle.v=new_v
    balle.theta=new_theta
end
```

```julia
function dt!(balle::Balle, partie::Partie)
    xp=balle.x+dt*v*cos(theta)
    yp=balle.y+dt*v*sin(theta)
    deplacement=LineSegment([balle.x,balle.y], [xp,yp])
    for i in Partie.obstacles
        if intersects(deplacement,i)
            Partie.is_over = true
        end
    end
    if !Partie.is_over
        balle.v+=dt*f/m
        balle.x=xp
        balle.y=yp
    end
end
```

```julia
function deplacement!(partie::Partie)
    while(partie.balle.v>0)
        dt!(partie)
    end
end
```

```julia
function isWon(partie::Partie)
    length = sqrt((partie.trou.x-partie.balle.x)^2+(partie.trou.y-partie.balle.y)^2)
    length<rt
end
```
